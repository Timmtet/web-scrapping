###  Youtube Tutorial

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
names = soup.find_all('a', class_='title')
print(names)
```

```python
product_name= []
for i in names:
    name = i.text
    product_name.append(name)
print(product_name)
```

```python
prices = soup.find_all('h4', class_= 'price')
print(prices)
```

```python
product_price = []
for i in prices:
    price = i.text
    product_price.append(price)
print(product_price)
```

```python
descs = soup.find_all('p', class_ = 'description')
print(descs)
```

```python
product_descs = []
for i in descs:
    desc = i.text
    product_descs.append(desc)
print(product_descs)
```

### Web scraping of tabular data

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://ticker.finology.in/'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'lxml')
```

```python
table = soup.find("table",  class_ = "table table-sm table-hover screenertable")
print(table)
```

```python
headers =table.find_all("th")
print(headers)
```

```python
header_list =[]
for i in headers:
    header= i.text
    header_list.append(header)
header_list
```

```python
df = pd.DataFrame(columns = header_list)
df
```

```python
rows = table.find_all('tr')
rows
```

```python
lines = []
for i in rows[1:]:
    data = i.text
    lines.append(data)
lines
```

```python
for i in rows[1:]:
    data = i.find_all('td')
    print(data)
```

```python
for i in rows[1:]:
    data = i.find_all('td')
    row = [tr.text for tr in data]
    l= len(df)
    df.loc[l] = row
    
print(df)
```
